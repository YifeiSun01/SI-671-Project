```python
import pandas as pd

df = pd.read_csv("data_lat_lng_390660.csv", index_col=0)
df["SALE DATE"] = pd.to_datetime(df["SALE DATE"])
df["YEAR BUILT"] = pd.to_datetime(df["YEAR BUILT"])
df
```

```python
df.dtypes
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a histogram of the date distribution
plt.figure(figsize=(15, 6))
plt.hist(df["SALE DATE"].dropna(), bins=500, color='skyblue', edgecolor='skyblue')
plt.title('Distribution of Dates')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
keep_list = []
for column in df.columns:
    if 1-np.sum(df[column].isna())/len(df) > 0.5:
        keep_list.append(column)
df = df[keep_list].dropna()
df.index = list(range(len(df)))
df
```

```python
import numpy as np

for column in df.columns:
    print(column,"    ",1-np.sum(df[column].isna())/len(df)," of the data is present")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a histogram of the date distribution
plt.figure(figsize=(15, 6))
plt.hist(df["SALE DATE"].dropna(), bins=300, color='skyblue', edgecolor='skyblue')
plt.title('Distribution of Dates')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.show()
```
